# Feature Diagnostics

This notebook reads the saved feature table and feature diagnostics generated by the pipeline.

It is the best place to inspect missing values, non-finite values, correlations, and selected features.

```python
import json
from pathlib import Path

import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()

features_df = pd.read_csv(ROOT / "data/processed/features_ltr.csv")
with open(ROOT / "data/experiments/feature_diagnostics.json", encoding="utf-8") as handle:
    diagnostics = json.load(handle)

pd.DataFrame(diagnostics["feature_summary"]).sort_values(["missing_count", "inf_count"], ascending=False), features_df.head()
```
